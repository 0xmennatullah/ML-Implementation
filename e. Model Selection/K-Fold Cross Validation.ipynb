```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
from sklearn import linear_model, tree, ensemble
```

```python
t = pd.read_csv('./data/house-prices-advanced-regression-techniques/train.csv')
t.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = t.SalePrice
t.drop(['SalePrice'], axis=1, inplace=True)
t.drop(['LotFrontage', 'GarageYrBlt', 'MasVnrArea'], axis=1, inplace=True)

numer_cols = [cname for cname in t.columns if t[cname].dtype in ['int64', 'float64']]

x = t[numer_cols].copy()
x.head()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cnt= 1
for idxr, idxs in kf.split(x,y):
    print(f'fold: {cnt}, train set: {len(idxr)}, test set: {len(idxs)}')
    cnt+=1
```

```python
def rmse(score):
    rmse = np.sqrt(-score)
    print(f'rmse = {'{:.2f}'.format(rmse)}')
```

```python
score = cross_val_score(linear_model.LinearRegression(), x, y, cv=kf, scoring="neg_mean_squared_error")
print(f'score: {score}')
rmse(score.mean())
```
